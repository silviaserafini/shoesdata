```python
def get_feature(data,values,default,func = lambda x:x):
    res = data

    try:
        for value in values:
            res = res[value]
        if res == 'null' or res == None:
            res = default
        return func(res)
        
    except:
        #print(f"{values} not available, returning {default}")
        pass
        return default
```
